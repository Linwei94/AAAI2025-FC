```python
import torch
import numpy
import torch.nn as nn
import torch.nn.functional as F
import matplotlib.pyplot as plt
```

```python
# load data
in_bin = torch.load('./in_bin.pth')
high_ce_indexs = torch.load('./high_ce_indexs.pth')
wrong_indexs_all = torch.load('./wrong_indexs.pth')
indexs = torch.load('./indexs.pth')
features = torch.load('./features.pth')
confidences = torch.load('./confidences.pth')

# check how many wrong_indexs_all in high_ce_indexs
wrong_indexs = []
for i in wrong_indexs_all:
    if i in high_ce_indexs:
        wrong_indexs.append(i)
print('wrong_indexs:', len(wrong_indexs))

# correct indexs
correct_indexs = []
for i in indexs:
    if i not in wrong_indexs:
        correct_indexs.append(i)
print('correct_indexs:', len(correct_indexs))
```

```python
wrong_mean_feature = features[wrong_indexs].mean(0).cpu()
correct_mean_feature = features[correct_indexs].mean(0).cpu()

# plot bar chart for correct_mean_feature
plt.figure(figsize=(20, 10))
plt.bar(range(0, 2048), correct_mean_feature)
plt.bar(range(0, 2048), wrong_mean_feature, alpha=0.5)
plt.legend(['correct', 'wrong'])
plt.show()
```
